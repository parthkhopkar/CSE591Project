```python
import matplotlib.pyplot as plt
```

```python
import pickle
```

```python
with open('./dynamic_lm_1_error_dd.pkl','rb') as f:
    dynamic_lm_1_error_dd = pickle.load(f)
```

```python
with open('./static_1_error.pkl','rb') as f:
    static_1_error = pickle.load(f)
```

```python
with open('./times.pkl','rb') as f:
    times = pickle.load(f)
```

```python
with open('./dynamic_lm_1_error_ndd.pkl','rb') as f:
    dynamic_lm_1_error_ndd = pickle.load(f)
```

```python
plt.plot(times,static_1_error,'-k', label='Static Landmarks')
plt.plot(times, dynamic_lm_1_error_ndd, '-r', label='Dynamic Landmarks without detection')
plt.plot(times, dynamic_lm_1_error_dd, '-g', label='Dynamic Landmarks with detection')
plt.ylabel('Error')
plt.xlabel('Time')
plt.title('Localization Errors')
plt.legend()
plt.savefig('1RobotCase_150dpi.png', dpi=150)
plt.show()
```

```python
with open('./dynamic_lm_2_error_dd_ms_robot_1_random.pkl','rb') as f:
    dynamic_lm_2_error_dd_ms_robot_1 = pickle.load(f)
with open('./dynamic_lm_2_error_dd_ms_robot_2_random.pkl','rb') as f:
    dynamic_lm_2_error_dd_ms_robot_2 = pickle.load(f)
```

```python
with open('./dynamic_lm_2_error_dd_no_ms_robot_1_random.pkl','rb') as f:
    dynamic_lm_2_error_dd_no_ms_robot_1 = pickle.load(f)
with open('./dynamic_lm_2_error_dd_no_ms_robot_2_random.pkl','rb') as f:
    dynamic_lm_2_error_dd_no_ms_robot_2 = pickle.load(f)
import numpy as np    
robot_avg_ms = (np.array(dynamic_lm_2_error_dd_ms_robot_1) + np.array(dynamic_lm_2_error_dd_ms_robot_2)) / 2
robot_avg_no_ms = (np.array(dynamic_lm_2_error_dd_no_ms_robot_1) + np.array(dynamic_lm_2_error_dd_no_ms_robot_2)) / 2
```

```python

plt.plot(times,robot_avg_ms,'-.k', label='Average error with Map Sharing')
plt.plot(times, robot_avg_no_ms, '-.b', label='Average error without Map Sharing')
#plt.plot(times, dynamic_lm_1_error_dd, '-g', label='Dynamic Landmarks with detection')
#plt.axis([0,120, 0, 6])
plt.ylabel('Error')
plt.xlabel('Time')
plt.title('Localization Errors (Map Sharing Experiments)')
plt.legend()
plt.savefig('AvgErrorRobotCase_300dpi.png', dpi=300)
plt.show()
```

```python
plt.plot(times,dynamic_lm_2_error_dd_ms_robot_1,'r', label='Two Robots with Map Sharing')
plt.plot(times,dynamic_lm_2_error_dd_no_ms_robot_1,'g', label='Single Robot')
plt.ylabel('Error')
plt.xlabel('Time')
plt.title('Localization Errors (Single Robot To Multi Robot)')
plt.legend()
plt.savefig('2RobotCase_300dpi.png', dpi=300)
plt.show()
```
